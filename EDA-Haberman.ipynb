# Exploratory data analysis on Haberman survival dataset

# Data Loading and Superficial Look

```python
# Importing necessary Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load Haberman data in to pandas
haberman = pd.read_csv('haberman.csv')
```

```python
# High level statistics of the dataset: 
#=================================================================================

# number of points, numer of features, number of classes, data-points per class.
print(haberman.shape)
```

```python
# => no of points = 305, no of features = 4
```

```python
# Giving Column Names to the dataset
haberman.columns = ['Age', 'Op_Year', 'axil_nodes', 'Surv_status']
```

```python
# => Column Names in the Dataset
print(haberman.columns)
```

```python
haberman.head()
```

```python
haberman['Surv_status'].value_counts()
```

```python
# => Number of Classes = 2, data-points per class ( 1 -> 224, 2 -> 81)
# Here :
# 1 = the patient survived 5 years or longer
# 2 = the patient died within 5 year

# It is an imbalanced dataset as the number of data points for both classes differ by a big difference.
```

```python
# => Objective -> Understand to differentiate two classes by closely understanding the features and data points associated with them.
```

# UNIVARIATE ANALYSIS

```python
# Data Categorization

haber_surv = haberman.loc[haberman["Surv_status"] == 1];
haber_dead = haberman.loc[haberman['Surv_status'] == 2];
```

```python
haber_surv.head()
```

```python
haber_dead.head()
```

# => 1-D Scatter Plot

```python
# 1-D Scatter Plot
#--------------------------------------------------
# Features - Age, Op_Year, axil_nodes

# 1-D Scatter Plot of Age
plt.plot(haber_surv["Age"], np.zeros_like(haber_surv["Age"]), 'o')
plt.plot(haber_dead["Age"], np.zeros_like(haber_dead["Age"]), 'o')
plt.show();
```

```python
# 1-D Scatter Plot of Op_Year
plt.plot(haber_surv["Op_Year"], np.zeros_like(haber_surv["Op_Year"]), 'o')
plt.plot(haber_dead["Op_Year"], np.zeros_like(haber_dead["Op_Year"]), 'o')
plt.show();
```

```python
# 1-D Scatter Plot of Op_Year
plt.plot(haber_surv["axil_nodes"], np.zeros_like(haber_surv["axil_nodes"]), 'o')
plt.plot(haber_dead["axil_nodes"], np.zeros_like(haber_dead["axil_nodes"]), 'o')
plt.show();
```

**Observations**
- Clearly using 1-D Scatter Plots for all the three features we cannot make any clear conclusions from data points on how theses classes can be classified.

# => Histogram/PDF

```python
# Histogram
#--------------------------------------------------
# Features - Age, Op_Year, axil_nodes

# Histogram of Age
plt.close();
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="Surv_status", size=5).map(sns.distplot, "Age").add_legend();
plt.show();
```

```python
# Histogram of Op_Year
plt.close();
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="Surv_status", size=5).map(sns.distplot, "Op_Year").add_legend();
plt.show();
```

```python
# Histogram of axil_nodes
plt.close();
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="Surv_status", size=5).map(sns.distplot, "axil_nodes").add_legend();
plt.show();
```

**Observations**
- We can clearly see that using Histograms for all the three features we cannot make any clear conclusions from data points on how theses classes can be distinguished.

# => CDF

```python
# CDF
#---------------------------------------------------------------------------------
# Features - Age, Op_Year, axil_nodes

# CDF of Age
counts, bin_edges = np.histogram(haber_surv['Age'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.figure(1)
plt.subplot(311)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(haber_dead['Age'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);
plt.show()

#---------------------------------------------------------------------------------
# CDF of Op_Year
counts, bin_edges = np.histogram(haber_surv['Op_Year'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.subplot(312)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(haber_dead['Op_Year'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);
plt.show()

#---------------------------------------------------------------------------------
# CDF of axil_nodes
counts, bin_edges = np.histogram(haber_surv['axil_nodes'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.subplot(313)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(haber_dead['axil_nodes'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf);
plt.plot(bin_edges[1:], cdf);
plt.show()
```

**Observations**
- We can clearly see that using CDF for all the three features we cannot make any clear conclusions.

```python
#Mean, Variance, Std-deviation
#----------------------------------------------------------------------------

# For Age
print('Means for Age:')
print(np.mean(haber_surv['Age']))
print(np.mean(haber_dead['Age']))

print("\nStd-dev for Age:");
print(np.std(haber_surv['Age']))
print(np.std(haber_dead['Age']))

# For Op_Year
print('Means for Op_Year:')
print(np.mean(haber_surv['Op_Year']))
print(np.mean(haber_dead['Op_Year']))

print("\nStd-dev for Op_Year:");
print(np.std(haber_surv['Op_Year']))
print(np.std(haber_dead['Op_Year']))

# For axil_nodes
print('Means for axil_nodes:')
print(np.mean(haber_surv['axil_nodes']))
print(np.mean(haber_dead['axil_nodes']))

print("\nStd-dev for axil_nodes:");
print(np.std(haber_surv['axil_nodes']))
print(np.std(haber_dead['axil_nodes']))
```

```python
#Median, Quantiles, Percentiles, IQR.

# For Age
print("\nMedians for Age:")
print(np.median(haber_surv['Age']))
print(np.median(haber_dead['Age']))

# For Op_Year
print("\nMedians for Op_Year:")
print(np.median(haber_surv['Op_Year']))
print(np.median(haber_dead['Op_Year']))

# For axil_nodes
print("\nMedians for axil_nodes:")
print(np.median(haber_surv['axil_nodes']))
print(np.median(haber_dead['axil_nodes']))
```

# BI-VARIATE ANALYSIS

# => 2-D Scatter Plot

```python
# Scatter Plots
#---------------------------------------------------------------------------------
# Features - Age, Op_Year, axil_nodes

sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="Surv_status", size=4).map(plt.scatter, "Age", "Op_Year").add_legend();
plt.show();
```

**Observations**
- Using Age and Op_Year features, we cannot distinguish patients of two different classes.

# => Pair-Plot

```python
# Pair Plots
#---------------------------------------------------------------------------------
# Features - Age, Op_Year, axil_nodes
plt.close();
sns.set_style("whitegrid");
sns.pairplot(haberman, hue="Surv_status", size=3);
plt.show()
```

**Observations**
- None of these features used pair-wise can clearly classify the data points in to two classes.
- All these data points of two classes have large overlaps and hence are not linearly seperable.
- Pair-plots can not be used to clearly identify how these two classes can be seperated.

# Conclusion:

- Objective of our problem was to classfiy data points in to different classes using a feature or combination of features.
- Using uni-variate analysis we cannot clearly find a means of classification.
- Using bi-variate analysis we cannot clearly find a means to classify. Data points of different classes have high overlaps and hence are not linearly seperable.
- We will have to find some other ways or high dimensional data analysis to find a way for classification.
